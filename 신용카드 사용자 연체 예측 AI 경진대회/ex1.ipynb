```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
train.head()
```

```python
le = LabelEncoder()
train["gender"] = le.fit_transform(train['gender'])
test["gender"] = le.fit_transform(test['gender'])
train.head()
```

```python
train["car"] = le.fit_transform(train['car'])
test["car"] = le.fit_transform(test['car'])
train.head()
```

```python
train["reality"] = le.fit_transform(train['reality'])
test["reality"] = le.fit_transform(test['reality'])
train.head()
```

```python
train["income_type"] = le.fit_transform(train['income_type'])
test["income_type"] = le.fit_transform(test['income_type'])
train.head()
```

```python
train["edu_type"] = le.fit_transform(train['edu_type'])
test["edu_type"] = le.fit_transform(test['edu_type'])
train.head()
```

```python
train["family_type"] = le.fit_transform(train['family_type'])
test["family_type"] = le.fit_transform(test['family_type'])
train.head()
```

```python
train["house_type"] = le.fit_transform(train['house_type'])
test["house_type"] = le.fit_transform(test['house_type'])
train.head()
```

```python
train = train.fillna({'occyp_type':'No job'})
test = test.fillna({'occyp_type':'No job'})
```

```python
train["occyp_type"] = le.fit_transform(train['occyp_type'])
test["occyp_type"] = le.fit_transform(test['occyp_type'])
train.head()
```

```python
train['Age'] = abs(train['DAYS_BIRTH'])/360
train = train.astype({'Age': 'int'})
test['Age'] = abs(test['DAYS_BIRTH'])/360
test = test.astype({'Age': 'int'})
```

```python
train.head()
```

```python
train['EMPLOYED']= train['DAYS_EMPLOYED'].map(lambda x: 0 if x>0 else x)
train['EMPLOYED'] = abs(train['EMPLOYED'])/360
train = train.astype({'EMPLOYED': 'int'})
test['EMPLOYED']= test['DAYS_EMPLOYED'].map(lambda x: 0 if x>0 else x)
test['EMPLOYED'] = abs(test['EMPLOYED'])/360
test = test.astype({'EMPLOYED': 'int'})
```

```python
train['Month'] = abs(train['begin_month'])
train = train.astype({'Month': 'int'})
test['Month'] = abs(test['begin_month'])
test = test.astype({'Month': 'int'})
```

```python
train = train.drop(['index','DAYS_BIRTH','DAYS_EMPLOYED','begin_month'],axis=1)
train.head()
```

```python
test = test.drop(['index','DAYS_BIRTH','DAYS_EMPLOYED','begin_month'],axis=1)
test.head()
```
